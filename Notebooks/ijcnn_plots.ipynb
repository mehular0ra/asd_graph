```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data = {
    "k": [100, 60, 40, 25, 5],
    "mean": [73.74, 75.33, 75.34, 74.39, 72.764],
    "ci_lower": [73.24, 74.78, 74.87, 72.67, 71.624],
    "ci_upper": [74.24, 75.88, 75.81, 76.11, 73.904],
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 3))
sns.set(style="white")
plt.plot(df['k'], df['mean'], color='green', marker='o', label='Mean')

# Adding the Confidence Interval
plt.fill_between(df['k'], df['ci_lower'], df['ci_upper'], color='lightgreen', alpha=0.3)

plt.xlabel("k Value")
plt.ylabel("Accuracy (%)")
plt.title("Number of ROIs in a hyperedge")
plt.xticks(df['k'])
# plt.legend()
plt.tight_layout()

# plt.show()

# Saving the plot
plt.savefig("k_value.png")
```

```python
data = {
    "L": [1, 2, 3, 4, 5],
    "mean": [75.34, 74.12, 74.39, 71.81, 66.8],
    "ci_lower": [74.87, 73.74, 74.39, 70.46, 65.46],
    "ci_upper": [75.81, 74.50, 74.39, 73.16, 68.14],
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 3))
sns.set(style="white")  # Turn off the grid
plt.plot(df['L'], df['mean'], marker='o', color='blue', label='Mean')  # Changed color to blue

# Adding the Confidence Interval
plt.fill_between(df['L'], df['ci_lower'], df['ci_upper'], color='lightblue', alpha=0.3)  # Light blue fill

plt.xlabel("Number of Layers")
plt.ylabel("Accuracy (%)")
plt.title("Number of Hypegraph convolution layers")
plt.xticks(df['L'])
plt.tight_layout()

# plt.show()

# Saving the plot
plt.savefig("L_value.png")

```
