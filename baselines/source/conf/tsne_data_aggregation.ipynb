```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import umap.umap_ as umap
```

```python

# DWHGN
# Define the base location
base_location = "/home/mehul/asd_graph/baselines/outputs/2023-09-03/11-31-08"

# Define the file paths using the base location
x_file_paths = [
    f"{base_location}/x_epoch_51_iter_{i}.npy" for i in range(6)
]

label_file_paths = [
    f"{base_location}/labels_epoch_51_iter_{i}.npy" for i in range(6)
]

# Load the x files
x_arrays = [np.load(file_path) for file_path in x_file_paths]

# Load the label files
label_arrays = [np.load(file_path) for file_path in label_file_paths]

# View the loaded arrays (for demonstration, we'll print the shapes)
for i, (x, label) in enumerate(zip(x_arrays, label_arrays)):
    print(f"x_{i} shape: {x.shape}")
    print(f"label_{i} shape: {label.shape}")
    print('---')

# Concatenate the arrays
# Concatenate arrays along the first axis
x_final = np.concatenate(x_arrays, axis=0)
label_final = np.concatenate(label_arrays, axis=0)

# Print the shapes of the final arrays
print(f"x_final shape: {x_final.shape}")
print(f"label_final shape: {label_final.shape}")

# Save the final arrays
# np.save(f"./DwHGN_x_final.npy", x_final)
# np.save(f"./DwHGN_label_final.npy", label_final)
# print('\nSaved the final arrays')
```

```python
# Apply UMAP to x_final
reducer = umap.UMAP(random_state=42)
x_umap = reducer.fit_transform(x_final)

# Plotting
plt.figure(figsize=(10, 7), dpi=100)  # Increase dpi for higher resolution
plt.style.use('seaborn-whitegrid')  # White grid style background

# Use scatter plot for the UMAP 2D representation
scatter = plt.scatter(x_umap[:, 0], x_umap[:, 1],
                      # Use 'viridis' as colormap for a more sophisticated look
                      c=label_final, cmap='viridis',
                      edgecolor='w', s=50, alpha=0.8)  # Increase marker size and set edgecolor to white for distinction

# Add colorbar to show the color coding based on labels
cbar = plt.colorbar(scatter, pad=0.01)  # Adjust padding for colorbar
cbar.set_label('Labels', size=14)  # Adjust font size for colorbar label
cbar.ax.tick_params(labelsize=12)  # Adjust font size for colorbar ticks

plt.title('UMAP Visualization of x_final', fontsize=18,
          pad=20)  # Adjust title font size and padding
plt.xlabel('UMAP component 1', fontsize=14)
plt.ylabel('UMAP component 2', fontsize=14)
plt.xticks(fontsize=12)  # Adjust x-tick font size
plt.yticks(fontsize=12)  # Adjust y-tick font size

# Remove right and top axis spines
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import os


def get_data(epoch, base_location):
    # Define the file paths using the base location
    x_file_paths = [
        f"{base_location}/x_epoch_{epoch}_iter_{i}.npy" for i in range(6)
    ]

    label_file_paths = [
        f"{base_location}/labels_epoch_{epoch}_iter_{i}.npy" for i in range(6)
    ]

    # Load the x files
    x_arrays = [np.load(file_path) for file_path in x_file_paths]

    # Load the label files
    label_arrays = [np.load(file_path) for file_path in label_file_paths]

    # View the loaded arrays (for demonstration, we'll print the shapes)
    for i, (x, label) in enumerate(zip(x_arrays, label_arrays)):
        # print(f"x_{i} shape: {x.shape}")
        # print(f"label_{i} shape: {label.shape}")
        # print('---')
        pass

    # Concatenate the arrays
    # Concatenate arrays along the first axis
    x_final = np.concatenate(x_arrays, axis=0)
    label_final = np.concatenate(label_arrays, axis=0)

    return x_final, label_final


def plot_data(x_final, label_final, base_location, epoch):
    # Apply UMAP to x_final
    reducer = umap.UMAP(random_state=42)
    x_umap = reducer.fit_transform(x_final)

    # Plotting
    plt.figure(figsize=(6, 4), dpi=100)  # Smaller figure size
    plt.style.use('seaborn-white')  # No grid, plain white background

    # Use scatter plot for the UMAP 2D representation
    label_classes = [0, 1]
    colors = ['blue', 'red']
    labels = ["TD", "ASD"]

    for i, (c, label) in enumerate(zip(colors, labels)):
        plt.scatter(x_umap[label_final == i, 0], x_umap[label_final == i, 1],
                    color=c, label=label, edgecolor='w', s=50, alpha=0.8)

    if base_location == '/home/mehul/asd_graph/baselines/outputs/2023-09-05/10-51-29':
        title_text = 'HyperGraphGCN'
    elif base_location == '/home/mehul/asd_graph/baselines/outputs/2023-09-03/11-31-08':
        title_text = 'GCN'
    elif base_location == '/home/mehul/asd_graph/baselines/outputs/2023-09-05/10-37-58':
        title_text = 'HyperGEL'
    plt.title(title_text, fontsize=16, pad=15)

    # Remove x and y tick values
    plt.xticks([])
    plt.yticks([])

    # Remove right and top axis spines
    ax = plt.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)

    # Legend
    leg = plt.legend(loc='upper right', fontsize=10,
                     frameon=True, edgecolor='black')
    for lh in leg.legendHandles:
        lh.set_alpha(1)  # Make legend symbols opaque

    # Save the plot
    # directory_name = os.path.basename('.')
    # save_directory = os.path.join(
    #     '/home/mehul/asd_graph/Notebooks', f"{directory_name}_plots")

    # if not os.path.exists(save_directory):
    #     os.makedirs(save_directory)  # Create directory if it doesn't exist

    # plt.savefig(f"{save_directory}/epoch_{epoch}.png", bbox_inches='tight')
    # plt.close()

    # Save in a 'plots' folder in the current directory
    save_directory = os.path.join(os.getcwd(), "plots")

    if not os.path.exists(save_directory):
        os.makedirs(save_directory)  # Create directory if it doesn't exist

    plt.savefig(f"{save_directory}/{title_text}_epoch_{epoch}.png",
                bbox_inches='tight')
    plt.close()



def get_data_and_plot(epoch, base_location):
    x_final, label_final = get_data(epoch, base_location)
    plot_data(x_final, label_final, base_location, epoch)
```
