### The below code can converts to hyperedge_index_matrix
Just change sub_id to get more subjects

```python
hyperedge_index = saved_tensors['sub2']['hyperedge_index'][0]
hyperedge_index = np.array(hyperedge_index)

# take out only those hyperedge_index pairs which have the hyperedge_index[1] value as <= 399
sub_hyperedge_index = hyperedge_index[:, hyperedge_index[1] <= 399]
print(sub_hyperedge_index.shape)


# make the hyperedge_index to a matrix of size 400 x 400, with 1s where there is a connection
# 1 if there is a connection, 0 if there is no connection, 
# hyperedge_index 0th row represents node, ie, row values
# hyperedge_index 1st row represents hyper_edge, ie, column values
# if there is (node, hyper_edge) pair in hyperedge_index, then put 1 in the matrix, else 0
sub_hyperedge_index_matrix = np.zeros((400, 400))
for i in range(len(sub_hyperedge_index[0])):
    sub_hyperedge_index_matrix[sub_hyperedge_index[0][i]][sub_hyperedge_index[1][i]] = 1
```

```python
print(sub_hyperedge_index_matrix.shape)
print(np.unique(sub_hyperedge_index_matrix, return_counts=True))
```

```python
filename = '/mnt/ssd1/mehul_data/research/Schaefer2018_400Parcels_7Networks_order.txt'
ROI_mapping = {}

with open(filename, 'r') as file:
    for line in file:
        # Split the line based on whitespace
        values = line.split()

        # Extract 'index' and 'name'
        index = int(values[0]) - 1
        name = values[1]

        # Remove the "7Networks_" prefix
        prefix = "7Networks_"
        if name.startswith(prefix):
            name = name[len(prefix):]

        # Add the cleaned name to the dictionary
        ROI_mapping[index] = name
```

```python
# Extracting the names
columnwise_roi_names = []

for col in range(sub_hyperedge_index_matrix.shape[1]):
    roi_names_for_col = [ROI_mapping[row] for row in range(
        sub_hyperedge_index_matrix.shape[0]) if sub_hyperedge_index_matrix[row, col] == 1]
    columnwise_roi_names.append(roi_names_for_col)

print(columnwise_roi_names)
print(len(columnwise_roi_names))
print(len(columnwise_roi_names[0]))
```

```python
from collections import Counter

columnwise_roi_frequencies = [Counter(sublist)
                              for sublist in columnwise_roi_names]

print(len(columnwise_roi_frequencies))
columnwise_roi_frequencies[0]['LH_Vis']
```
